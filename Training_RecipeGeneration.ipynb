```python
import numpy as np
import pickle
```

```python
chunk_size = 100000
```

```python
all_embeddings = []
for start in range(0, len(df), chunk_size):
    end = min(start + chunk_size, len(df))
    print(f"Encoding rows {start} to {end}...")
    
    chunk_text = df["ingredients_text"].iloc[start:end].tolist()
    
    chunk_emb = lightmodel.encode(
        chunk_text,
        batch_size=64,
        convert_to_tensor=False,
        show_progress_bar=True,
        device="cpu"
    )
    all_embeddings.extend(chunk_emb)

all_embeddings = np.array(all_embeddings)
```

```python
np.save("ingredients_embeddings.npy", all_embeddings)
print("saved ingredients")
```

```python
import numpy as np
from sentence_transformers import util
```

```python
emb = np.load("ingredients_embeddings.npy")
```

```python
def get_similar_recipes(user_query, top_k=1):
    user_emb = lightmodel.encode(user_query, convert_to_tensor=True)
    scores = util.cos_sim(user_emb, emb)[0].cpu().numpy()

    top_idx = scores.argsort()[-top_k:][::-1]

    results = df.iloc[top_idx].copy()
    results["similarity"] = scores[top_idx]
    return results
```

```python
inp = input("Enter the ingredients :" )
```

```python
results = get_similar_recipes(inp)
```

```python
print(results["steps"])
```
